# Clustering

```python
df_cluster = df_scaled.copy()
```

```python

sse = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, n_init=10, random_state=42)  # Change 'auto' to an integer value, e.g., 10
    kmeans.fit(df_cluster)
    sse.append(kmeans.inertia_)

plt.plot(range(1, 11), sse, marker='o', linestyle='-')
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.show()

findknee = KneeLocator(range(1, 11),
                       sse,
                       curve="convex",
                       direction="decreasing")

print('best cluster:', findknee.elbow)
```

```python
best_kmeans = KMeans(n_clusters=findknee.elbow, n_init=5, random_state=42)  
best_kmeans.fit(df_cluster)
df['Kmeans Cluster'] = best_kmeans.labels_
```

```python
aggregation_dic = {
    'Marital status': 'mean',  
    'Application mode': 'mean',  
    'Previous qualification': 'mean', 
    'Admission grade': 'mean',
    'Tuition fees up to date': lambda x: x.mode().iloc[0],
    'Scholarship holder': lambda x: x.mode().iloc[0],
    'Age at enrollment': 'mean', 
    'Curricular units 1st sem (approved)': 'mean',
    'Curricular units 1st sem (grade)': 'mean', 
    'Curricular units 2nd sem (approved)': 'mean',
    'Curricular units 2nd sem (grade)': 'mean', 
    'GDP': 'mean', 
    'Target': lambda x: x.mode().iloc[0]  
}
summary = df.groupby('Kmeans Cluster').agg(aggregation_dic)
summary
```

```python
hch= AgglomerativeClustering(n_clusters=findknee.elbow)
y_hch = hch.fit_predict(df_scaled)


df["Hierarchical Cluster"] = y_hch
aggregation_dict = {
    'Marital status': 'mean',  
    'Application mode': 'mean',  
    'Previous qualification': 'mean', 
    'Admission grade': 'mean',
    'Tuition fees up to date': lambda x: x.mode().iloc[0],
    'Scholarship holder': lambda x: x.mode().iloc[0],
    'Age at enrollment': 'mean', 
    'Curricular units 1st sem (approved)': 'mean',
    'Curricular units 1st sem (grade)': 'mean', 
    'Curricular units 2nd sem (approved)': 'mean',
    'Curricular units 2nd sem (grade)': 'mean', 
    'GDP': 'mean', 
    'Target': lambda x: x.mode().iloc[0]  
}
summary = df.groupby('Hierarchical Cluster').agg(aggregation_dict)
summary
```

# Classification

```python
df_class = df_scaled.copy()
```

```python
X = df_class.drop(['Target'], axis=1)
y = df_class['Target']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)
rf_pred = rf_model.predict(X_test)
```

```python
actualData = label_encoder.inverse_transform(y_test)
predicted_labels = label_encoder.inverse_transform(rf_pred)

comparison_df = pd.DataFrame({'Actual': actualData, 'Predicted': predicted_labels})

rf_accuracy = accuracy_score(y_test, rf_pred)  
conf_matrix_rf = confusion_matrix(y_test, rf_pred)
print("Confusion Matrix:")
print(conf_matrix_rf)
print(f"Random Forest Accuracy: {rf_accuracy:.4f}")
comparison_df.sample(n=10)
```

```python
logisModel = LogisticRegression(multi_class='multinomial', solver='lbfgs', max_iter=1000)
logisModel.fit(X_train, y_train)
logisPred = logisModel.predict(X_test)

log_reg_accuracy = accuracy_score(y_test, logisPred)
conf_matrix = confusion_matrix(y_test, logisPred)
print("Confusion Matrix:")
print(conf_matrix)
print(f"Accuracy: {log_reg_accuracy:.4f}")

actual_labels = label_encoder.inverse_transform(y_test)
predicted_labels = label_encoder.inverse_transform(logisPred)

comparison_df = pd.DataFrame({'Actual': actual_labels, 'Predicted': predicted_labels})
comparison_df.sample(n=10)
```

```python
dtModel = DecisionTreeClassifier(random_state=42)
dtModel.fit(X_train, y_train)
dtPred = dtModel.predict(X_test)

tree_accuracy = accuracy_score(y_test, dtPred)
conf_matrix = confusion_matrix(y_test, dtPred)
print("Confusion Matrix:")
print(conf_matrix)
print(f"Accuracy: {tree_accuracy:.4f}")

actual_labels = label_encoder.inverse_transform(y_test)
predicted_labels = label_encoder.inverse_transform(dtPred)

comparison_df = pd.DataFrame({'Actual': actual_labels, 'Predicted': predicted_labels})
comparison_df.sample(n=10) 
```

```python
akurasi = [log_reg_accuracy, tree_accuracy, rf_accuracy]
models = ['Logistic Regression', 'Decision Tree', 'Random Forest']

plt.figure(figsize=(8, 6))
plt.bar(models, akurasi, color=['blue', 'green', 'red'])
for i, acc in enumerate(akurasi):
    plt.text(i, acc + 0.01, f'{acc:.3f}', ha='center', color='black', fontsize=10)
plt.title('Comparison of Classifier Accuracy')
plt.ylim(0, 1)
plt.show()
```
